Find the derivative of the function $f(x)=\sqrt{2x-1}$ from the definition

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
k=1e-5 
x = np.linspace(0.6,10,50)
y=[]
for i in x:	
	l=np.sqrt(2*(i+k)-1)-np.sqrt(2*i-1)
	l=l/k
	y.append(l)
fx=1/np.sqrt(2*x-1)
plt.plot(x,y,'o')
plt.plot(x,fx)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
# # # #Comment the following line
#plt.savefig('../figs/1.eps')
plt.show()
```

\begin{equation}
f(x)=
 \begin{cases} 
      \sin(\pi x) & 0<x < 1 \\
      \ln (x) & 1< x< 2  
   \end{cases},x_0=1
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
n = np.arange(0,2,1e-3)
b=[]
for i in n:
   	if(i<=1):
   		b.append(np.sin(np.pi*i))
   	else:
   		b.append(np.log(i))
plt.plot(n,b)
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
#plt.savefig('../figs/4.eps')
plt.show()
```
